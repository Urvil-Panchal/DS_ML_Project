# Data Transformation

```python
# First we will extract some data from one column
def funct(x):
    return x+10
    
df_mpg = df[['mpg']]
df_mpg = df_mpg.iloc[0:7]
df_mpg

# Then we will apply the function to that dataframe
df_mpg
```

```python
# Applying the function 

df_mpg['mpg'] = df_mpg['mpg'].apply(funct)

df_mpg
```

# apply() used apply function to single column
# applymap() is used to apply the function to whole data frame

# Data Aggregation: Used to perform aggregation function (mean , max , sum , std.. etc)on grouped data

```python
df
```

```python
# Grouping by number of cylinders

df_cyl = df.groupby('cylinders')
```

```python
df_cyl.agg({'mpg' : 'sum'})
```

```python
df_cyl.agg({'horsepower' : 'max'})
```

# Featuer Transformer: Label Encoder

Recommended only when the number of categories are less than or equals to 3.

```python
unique_origin_value = df['origin'].unique()
unique_origin_value
```

```python
# Label Encoder for Origin Column

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['origin'] = le.fit_transform(df['origin'])

df
```

```python
print('Before applying LabelEncoder the Origin values are:' , unique_origin_value)
print('After applying LabelEncoder the Origin values are',df.origin.unique())
```

# We will apply One_Hot Encoder for the cylinders Column

Recommended only when the number of categories are more than 3

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()

df_ohe = ohe.fit_transform(df[['cylinders']])

#Convert the encoded result to a DataFrame
df_encoded_new = pd.DataFrame(df_ohe.toarray(), columns=ohe.get_feature_names_out(['cylinders'])).astype(int)

df_encoded_new
```
