# Data Cleaning and Handling Missing Values:



```python
# There is no missing values in our dataset
# so for that we will make new data set having some missing values

data = {
    'key1' : [1,2,3,4,5],
    'key2' : [6,7,8,9,10],
    'key3' : [0,0,0,0,0],
    'key4' : ['Urvil' , 'Abhay' , 'Harsh' ,  'Jenil' , 'Darshan']
}

df_ = pd.DataFrame(data , index=['a' , 'b' , 'c' , 'd' , 'e'])
df_
```

```python
# Firstly we replace the key3 cols rows with NaN
import numpy as np
df_['key3'] = df_['key3'].replace(0,np.nan)
```

```python
# To check the missing values in cols
df_.isnull().sum()
```

```python
# To fill the missing values in place of NaN

df_.fillna('😎' , inplace=True)
```

```python
df_
```

```python
# We can also drop the whole column or entire row

df_.drop('key3',axis=1 , inplace=True)
```

# Data Merging and Joining:

```python
data1 = {
    'ID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 22, 27]
}

data2 = {
    'ID': [3, 4, 5, 6],
    'Department': ['HR', 'Finance', 'IT', 'Marketing'],
    'Salary': [50000, 60000, 45000, 55000]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
```

```python
merged_inner = pd.merge(df1, df2, on='ID', how='inner')
merged_outer = pd.merge(df1, df2, on='ID', how='outer')

# Joining DataFrames based on the index
joined_left = df1.set_index('ID').join(df2.set_index('ID'), how='left')
joined_right = df1.set_index('ID').join(df2.set_index('ID'), how='right')
```

# Data Reshaping:

```python
#Pivot Table

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Alice', 'Bob', 'Eva'],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR', 'Finance', 'IT', 'Marketing'],
    'Salary': [50000, 60000, 45000, 55000, 70000, 52000, 58000, 72000]
}

df_data = pd.DataFrame(data)

# Create a pivot table to calculate the average salary for each department
pivot_table = pd.pivot_table(df_data, values='Salary', index='Department', aggfunc='mean')

pivot_table
```

```python
# Pivot

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Alice', 'Bob', 'Eva'],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR', 'Finance', 'IT', 'Marketing'],
    'Salary': [50000, 60000, 45000, 55000, 70000, 52000, 58000, 72000]
}

df_data = pd.DataFrame(data)

# Create a pivot table using pivot() to reshape the data
pivot_table = df_data.pivot(index='Name', columns='Department', values='Salary')

pivot_table
```
